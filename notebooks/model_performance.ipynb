# Checking the Model's Performance

```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Load the model and the data
model = joblib.load('../model/spam_classifier_model.joblib')

X_test = pd.read_csv('../data/X.csv')
y_test = pd.read_csv('../data/y.csv')

# Convert the data to numpy arrays
y_test = y_test.values.ravel()
```

```python
# Predict the labels
y_pred = model.predict(X_test)
```

## Confusion Matrix

```python
matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['Not Spam', 'Spam'], yticklabels=['Not Spam', 'Spam'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## Classifier Report

```python
print("Classification Report:\n")
print(classification_report(y_test, y_pred))
```

## Bar Plot

```python
sns.countplot(x=y_pred, palette='viridis')
plt.title('Predicted Labels')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not Spam', 'Spam'])
plt.show()
```

## Comparing Real x Predict

```python
df_results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

sns.countplot(data=df_results,
              x='Actual',
              hue='Predicted',
              palette='Set2')
plt.title('Actual vs Predicted Labels')

plt.show()
```

Note: In green, the model predicted as "Not Spam" while in orange the model predicted as "Spam"
